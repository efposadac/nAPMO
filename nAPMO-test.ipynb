### Numerical integration test

```python
import utilities.numerical_integration as nint
```

Computes the integral of f(x) by using Gauss-Chebyshev quadrature of second kind.

Perez-Jorda, J., San-Fabian, E. & Moscardo, F. A simple, reliable and efficient scheme for automatic numerical integration. Comput. Phys. Commun. 70, 271-284 (1992).

Iterative method: The number of quadrature points increase until convergence is reached.
Return the integral value and the estimated error.

```python
f = lambda x: np.exp(-6.793 * x * x) /(1.000001 - (x*x)) # exact : 0.762416793289 
nint.chebgauss_integrate(f)
```

```python
f = lambda x: 1/(x**4 + x**2 + 0.9) # exact : 1.582232963730
nint.chebgauss_integrate(f)
```

### Grids test

```python
from interfaces.becke_grid import *
angularPoints = 590
radialPoints = 15
grid = BeckeGrid(radialPoints, angularPoints)
```

Radial Distribution

```python
for p in xrange(5,radialPoints):
    x, y = grid.radial_distr(p)
    title('Transformed Gauss-Chebyshev points mapped for radial integration')
    plot(x, (x*0)+p, 'o')
```

Angular distribution

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title('Lebedev quadrature points', loc='left')

data = grid.angular_distr(angularPoints)

ax.scatter(data[0], data[1], data[2], s=5, color='b')
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])
```
