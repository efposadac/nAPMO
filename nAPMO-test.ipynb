### Grids test

```python
angularPoints = 86
radialPoints = 15
grid = BeckeGrid(radialPoints, angularPoints)
```

Radial Distribution

```python
for p in xrange(5,radialPoints):
    x, y = grid.radial_distr(p)
    title('Transformed Gauss-Chebyshev points mapped for radial integration')
    plot(x, (x*0)+p, 'o')
```

Angular distribution

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title('Lebedev quadrature points', loc='left')

data = grid.angular_distr(angularPoints)

ax.scatter(data[0], data[1], data[2], s=5, color='b')
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])
```
